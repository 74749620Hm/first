RFM analysis

```python
# last transaction date
df["Order Date"].max()
```

```python
df.columns
```

```python


# Tentukan tanggal referensi (misalnya, hari terakhir pada dataset)
reference_date = df['Order Date'].max()

# Hitung nilai RFM untuk setiap pelanggan
rfm = df.groupby('Customer ID').agg({
    'Order Date': lambda x: (reference_date - x.max()).days,  # Recency
    'Order ID': 'nunique',                                    # Frequency
    'Sales': 'sum'                                            # Monetary
}).reset_index()

# Ubah nama kolom
rfm.columns = ['Customer ID', 'Recency', 'Frequency', 'Monetary']

# Buat scoring untuk masing-masing RFM
rfm['R_Score'] = pd.qcut(rfm['Recency'], 4, labels=[4, 3, 2, 1])  # Recency, semakin rendah semakin baik
rfm['F_Score'] = pd.qcut(rfm['Frequency'], 4, labels=[1, 2, 3, 4])  # Frequency, semakin tinggi semakin baik
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 4, labels=[1, 2, 3, 4])  # Monetary, semakin tinggi semakin baik

# Kombinasikan RFM Score
rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)

# Tentukan segmentasi berdasarkan RFM Score
def segment_rfm(df):
    if df['RFM_Score'] == '144':
        return 'Best Customers'
    elif df['R_Score'] == 4:
        return 'Loyal Customers'
    elif df['R_Score'] == 1:
        return 'At Risk'
    else:
        return 'Others'

rfm['Segment'] = rfm.apply(segment_rfm, axis=1)

# Tampilkan hasil RFM
print(rfm.head())
```

```python
rfm[rfm['Segment']=='Others']
```

```python
# Count plot untuk segmentasi
plt.figure(figsize=(8, 6))
sns.countplot(x='Segment', data=rfm, palette='Set2', order=rfm['Segment'].value_counts().index)
plt.title('Customer Segmentation Based on RFM', fontsize=14)
plt.xlabel('Segment')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Frequency', y='Monetary', hue='Segment', data=rfm, palette='Set1', s=100)
plt.title('Frequency vs Monetary with Customer Segments', fontsize=14)
plt.xlabel('Frequency')
plt.ylabel('Monetary')
plt.legend(title='Segment', loc='upper right')
plt.show()
```

```python
rfm_heatmap = rfm.pivot_table(index='R_Score', columns='F_Score', values='M_Score', aggfunc='mean')

plt.figure(figsize=(8, 6))
sns.heatmap(rfm_heatmap, annot=True, fmt=".1f", cmap="coolwarm", linewidths=.5)
plt.title('RFM Score Heatmap')
plt.show()
```

```python
# Set style
sns.set(style='whitegrid')

# Plotting Recency, Frequency, and Monetary distribution
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Recency distribution
sns.histplot(rfm['Recency'], bins=20, ax=axes[0], kde=True, color='skyblue')
axes[0].set_title('Recency Distribution')

# Frequency distribution
sns.histplot(rfm['Frequency'], bins=20, ax=axes[1], kde=True, color='lightgreen')
axes[1].set_title('Frequency Distribution')

# Monetary distribution
sns.histplot(rfm['Monetary'], bins=20, ax=axes[2], kde=True, color='salmon')
axes[2].set_title('Monetary Distribution')

plt.tight_layout()
plt.show()
```
